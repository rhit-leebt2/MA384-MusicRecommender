## Imports

```python
import pandas as pd
```

## Data Loading

```python
songs = pd.read_csv('./../data/dataset.csv')
songs.info()
```

### Data Exploration

```python
## how many unique genres are in the dataset?
print(f"Unique genres in the dataset: {len(songs.track_genre.value_counts())}")

## make genres categorical datatype to save memory
songs['track_genre'] = songs['track_genre'].astype('category')

## make track_id as string datatype
songs['track_id'] = songs['track_id'].astype('string')
songs['album_name'] = songs['album_name'].astype('string')
songs['track_name'] = songs['track_name'].astype('string')
```

```python
songs.groupby('track_genre')['liveness'].mean().sort_values(ascending=False).head().plot.bar(y='track_popularity', legend=False, figsize=(12,6), title='Average Track Popularity by Genre')
```

```python
songs.groupby('track_genre')['liveness'].mean().sort_values(ascending=True).head().plot.bar(y='track_popularity', legend=False, figsize=(12,6), title='Average Track Popularity by Genre')
```
