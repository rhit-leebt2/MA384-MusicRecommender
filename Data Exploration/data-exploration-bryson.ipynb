## Correlation Analysis

We will analyze correlations between numerical features to identify relationships that could be useful for music recommendation. Features like danceability, energy, valence, etc., may correlate with user preferences or song popularity.

```python
numerical_cols = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 
                  'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

From this heatmap, we can see that loudness/energy, and valence/danceability have the highest correlations, while acousticness/energy, and acousticness/loudness have the least. Thus, since acousticness seems negatively correlated with many features, it may be reasonable not to use it.

```python
corr_with_popularity = corr_matrix['popularity'].sort_values(ascending=False)
print("Correlation with Popularity:")
print(corr_with_popularity)
```

On first attempt, I would guess that popularity may have a correlation with other features, howevre it seems like there are not many correlated features.

## Very Basic Data Cleaning

Perform an initial data cleaning, removing null values, outliers, and unneeded columns. Unnamed: 0, mode, and track_id are both useless for prediction, so I will drop them first.

```python
cols_to_drop = ["Unnamed: 0", "track_id", "mode"]
cleaned_df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
cleaned_df = cleaned_df.dropna()

numeric_cols = cleaned_df.select_dtypes(include=[np.number]).columns

Q1 = cleaned_df[numeric_cols].quantile(0.25)
Q3 = cleaned_df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

df_clean = cleaned_df[~((cleaned_df[numeric_cols] < (Q1 - 1.5 * IQR)) |
                (cleaned_df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
df_clean.reset_index(drop=True, inplace=True)

df_clean.head()
```

## Recommend Song Based on Another Song

A primary use case that I determined would be for someone to want to find new music based on music they already like. This serves as a basic proof of concept for computing similarity. Future work should involve determining an accuracy metric and training a model that determines how weighted each feature should be for calclating similarity. I first define a basic KNN with no feature engineering:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

feature_cols = [
    "danceability", "energy", "loudness", "speechiness",
    "acousticness", "instrumentalness", "liveness",
    "valence", "tempo", "duration_ms"
]

X = df_clean[feature_cols]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

knn = NearestNeighbors(n_neighbors=2, metric="euclidean")
knn.fit(X_scaled)
```

Pick a random song and give the nearest song according to our KNN.

```python
random_index = np.random.randint(0, len(df_clean))
random_song = df_clean.iloc[random_index]

song_vector = scaler.transform([random_song[feature_cols]])

distances, indices = knn.kneighbors(song_vector)

nearest_index = indices[0][1]
nearest_song = df_clean.iloc[nearest_index]

display(random_song)
display(nearest_song)
```
