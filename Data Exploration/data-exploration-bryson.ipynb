## Correlation Analysis

We will analyze correlations between numerical features to identify relationships that could be useful for music recommendation. Features like danceability, energy, valence, etc., may correlate with user preferences or song popularity.

```python
numerical_cols = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 
                  'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

From this heatmap, we can see that loudness/energy, and valence/danceability have the highest correlations, while acousticness/energy, and acousticness/loudness have the least. Thus, since acousticness seems negatively correlated with many features, it may be reasonable not to use it.

```python
corr_with_popularity = corr_matrix['popularity'].sort_values(ascending=False)
print("Correlation with Popularity:")
print(corr_with_popularity)
```

On first attempt, I would guess that popularity may have a correlation with other features, howevre it seems like there are not many correlated features.

## Genre-Based Analysis

```python
genre_popularity = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False)
print("Top 10 Genres by Average Popularity:")
print(genre_popularity.head(10))

plt.figure(figsize=(12, 6))
genre_popularity.head(10).plot(kind='bar')
plt.title('Top 10 Genres by Average Popularity')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
top_genres = genre_popularity.head(5).index
genre_features = df[df['track_genre'].isin(top_genres)].groupby('track_genre')[numerical_cols].mean()
print("Average Feature Values for Top 5 Genres:")
print(genre_features)
```

## Insights

By looking at the data, the conclusions that I have come to are:
